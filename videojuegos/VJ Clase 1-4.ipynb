### 2. ¿Qué es un juego? Los 4 Pilares
Se descompone la idea abstracta de "juego" en componentes tangibles.

#### Pilar 1: Objetivo Claro 🎯

    Todo juego necesita una meta. Puede ser simple (llegar a 1000 puntos) o compleja (resolver un misterio).

    Ejemplos: En Pac-Man, el objetivo es comer todos los puntos. En Tetris, sobrevivir el mayor tiempo posible.

#### Pilar 2: Reglas y Limitaciones 📜

    Las reglas definen CÓMO se puede (y no se puede) alcanzar el objetivo. Las limitaciones crean el desafío.

    Ejemplos: En Arkanoid, la pala solo se mueve horizontalmente. En Space Invaders, solo puedes disparar un láser a la vez.

#### Pilar 3: Conflicto y Desafío 🔥

    Es el obstáculo entre el jugador y el objetivo. Sin desafío, no hay juego, hay un paseo.

    Ejemplos: Los fantasmas en Pac-Man, la velocidad creciente de las piezas en Tetris, los enemigos en Super Mario.

#### Pilar 4: Interacción y Retroalimentación (Feedback) 🎮
    Interacción: Lo que el jugador HACE (mover, saltar, disparar).

    Retroalimentación: Cómo el juego RESPONDE. Es crucial. Un sonido al agarrar una moneda, la pantalla que tiembla con una explosión, la puntuación que sube. Sin feedback, el juego se siente "muerto".

2. El Bucle Principal: El Corazón del Juego
El while ejecutando: es el Game Loop. Un juego no es un programa que corre y termina; es un ciclo que se repite unas 60 veces por segundo. Cada "vuelta" o "fotograma" del bucle tiene tres fases:

        INPUT (Entradas): Revisa si el jugador ha hecho algo (mover el ratón, pulsar una tecla, cerrar la ventana).

        UPDATE (Actualización): Actualiza el estado del juego (mover enemigos, comprobar colisiones, etc.). Por ahora, esta parte estará vacía.

        RENDER (Dibujo): Dibuja todo en la pantalla en su nueva posición.

Mejoraremos nuestro bucle para que siga esta estructura:



    # ... (código anterior de inicialización) ...

    # Variables para el bucle
    ejecutando = True

    while ejecutando:
        # 1. INPUT: Recorremos la lista de eventos
        for evento in pygame.event.get():
            # Si el evento es 'cerrar la ventana', salimos del bucle
            if evento.type == pygame.QUIT:
                ejecutando = False

        # 2. UPDATE: (Aquí irá la lógica del juego en el futuro)
        # Por ahora no hacemos nada

        # 3. RENDER: Dibujamos en la pantalla
        # Rellenamos el fondo de un color (RGB: Rojo, Verde, Azul)
        PANTALLA.fill((30, 30, 40)) # Un color azul oscuro

        # ¡Importante! Actualizamos la pantalla para que se muestren los cambios
        pygame.display.flip()

    # Salir de pygame y del programa
    pygame.quit()

        Un Objeto (o instancia) es la cosa real construida a partir del plano. Es la silla específica que construiste. Puedes usar el mismo plano para construir muchas sillas (muchos objetos).

## --- NUESTRA PRIMERA CLASE ---
    #creamos paquete core y dentro ponemos la clase player.py
    import pygame

    class Player:
        # El "constructor": se ejecuta UNA VEZ al crear el objeto
        def __init__(self, x, y, ancho, alto, color):
            # self se refiere al objeto específico que se está creando
            self.rect = pygame.Rect(x, y, ancho, alto)
            self.color = color
            self.velocidad = 5

        # Un "método": una función que pertenece a la clase
        def mover(self, teclas):
            if teclas[pygame.K_LEFT]:
                self.rect.x -= self.velocidad
            if teclas[pygame.K_RIGHT]:
                self.rect.x += self.velocidad
            if teclas[pygame.K_UP]:
                self.rect.y -= self.velocidad
            if teclas[pygame.K_DOWN]:
                self.rect.y += self.velocidad
        
        def dibujar(self, superficie):
            pygame.draw.rect(superficie, self.color, self.rect)

Explicación de la Sintaxis:

class Player:: Así declaramos una nueva clase.

    def __init__(self, ...):: Este es el método constructor. Se llama automáticamente cuando creas un objeto (jugador = Player(...)). Su trabajo es configurar las variables iniciales.

    self: Es la palabra clave más importante. Dentro de la clase, self se refiere al objeto mismo. self.rect es la "caja de colisiones del jugador", self.velocidad es "la velocidad del jugador". Permite que cada objeto Player que creemos en el futuro tenga sus propias variables.

    mover(self, teclas) y dibujar(self, superficie): Son métodos, es decir, funciones que el objeto "sabe hacer". Hemos encapsulado la lógica del movimiento y del dibujo dentro del propio objeto.

Parte 3: Pulido del Motor - Control del Tiempo

    Hay un problema grave en nuestro código actual. El jugador se mueve 5 píxeles por fotograma. Si el juego corre en un ordenador muy rápido (ej: 300 fotogramas por segundo), el jugador se moverá rapidísimo. Si corre en uno lento (30 FPS), se moverá como una tortuga.

    ¡El movimiento del juego debe depender del tiempo real, no de la velocidad de la PC!

La Solución: pygame.time.Clock
Pygame nos da un reloj para controlar esto.

Limitar los FPS: Podemos decirle a Pygame que no corra a más de, por ejemplo, 60 fotogramas por segundo. Esto nos da consistencia.

Movimiento basado en Delta Time (dt): Podemos medir cuántos milisegundos han pasado desde el último fotograma. A esto se le llama "delta time" o dt. Luego, multiplicamos nuestra velocidad por este dt. Así, el movimiento será el mismo sin importar los FPS.

Vamos a implementar la versión con límite de FPS, que es más sencilla para empezar.


    # ... (código de la clase Player, etc.) ...

    # --- ANTES DEL BUCLE PRINCIPAL ---
    jugador = Player(ANCHO / 2 - 25, ALTO / 2 - 25, 50, 50, ROJO)
    reloj = pygame.time.Clock() # Creamos un objeto Clock

    # --- BUCLE PRINCIPAL ---
    ejecutando = True
    while ejecutando:
        # ... (sección de INPUT y UPDATE son iguales) ...
        # jugador.mover(teclas)

        # --- RENDER ---
        PANTALLA.fill(AZUL_OSCURO)
        jugador.dibujar(PANTALLA)
        pygame.display.flip()

        # --- CONTROL DE TIEMPO ---
        # Le decimos al reloj que espere lo necesario para que el bucle
        # no se ejecute más de 60 veces por segundo.
        reloj.tick(60)
    pygame.quit()



Explicación:

    reloj = pygame.time.Clock(): Creamos el objeto reloj una sola vez, antes del bucle.

    reloj.tick(60): Esta línea, al final del bucle, hace dos cosas:

      Calcula cuánto tiempo ha pasado desde la última vez que se llamó.

      Hace una pausa para asegurar que el bucle completo no tarde menos de 1/60 de segundo. En la práctica, esto limita el juego a 60 FPS.

    class Collectible:
        # El constructor. Similar al del jugador, pero sin velocidad.
        def __init__(self, x, y, ancho, alto, color):
            self.rect = pygame.Rect(x, y, ancho, alto)
            self.color = color

        # El método para dibujarlo en pantalla
        def dibujar(self, superficie):
            pygame.draw.rect(superficie, self.color, self.rect)

### Parte 2: Detección de Colisiones
1. La Magia de colliderect
¿Cómo sabemos si el jugador y la moneda se están tocando? Pygame lo hace increíblemente fácil.

Recordemos que tanto jugador como moneda tienen un self.rect.

Los objetos Rect de Pygame tienen un método llamado colliderect().

jugador.rect.colliderect(moneda.rect) devolverá True si los dos rectángulos se están solapando, y False si no.

### Explicación:

import random: Trae las herramientas de aleatoriedad de Python.

puntuacion = 0: Inicializamos el marcador.

puntuacion += 1: Es una forma abreviada de escribir puntuacion = puntuacion + 1.

random.randint(min, max): Nos da un número entero aleatorio entre min y max (ambos incluidos). Lo usamos para darle a la moneda una nueva coordenada X e Y dentro de los límites de la pantalla.

2. Mostrando la Puntuación en Pantalla

Para mostrar texto en Pygame, seguimos 3 pasos:

*   Crear una fuente: Definir el tipo de letra y el tamaño.
*   Renderizar el texto: Crear una imagen (una Surface) a partir de un texto y un color.
*  Dibujar la imagen: Copiar esa imagen de texto a nuestra pantalla principal.





    # 1. Importar la librería de pygame
    import pygame

    from core.collectible import Collectible
    from core.player import Player
    from utils.colors import ROJO, AZUL_OSCURO, BLANCO
    import random

    # 2. Inicializar pygame
    pygame.init()

    # 3. Definir el tamaño de la ventana
    ANCHO = 800
    ALTO = 600
    PANTALLA = pygame.display.set_mode((ANCHO, ALTO))


    # 4. Ponerle un título a la ventana
    pygame.display.set_caption('Segundo Juego')

    # --- CREAMOS EL OBJETO JUGADOR ---
    jugador = Player(ANCHO / 2 - 25, ALTO / 2 - 25, 50, 50, ROJO)
    fuente = pygame.font.Font(None, 36)
    # Creamos una instancia de nuestra nueva clase
    moneda = Collectible(100, 100, 30, 30, (255, 223, 0))  # Un color dorado
    reloj = pygame.time.Clock()  # Creamos un objeto Clock
    puntuacion = 0  # Nuestra variable para la puntuación

    # --- BUCLE PRINCIPAL ---
    ejecutando = True
    while ejecutando:
        # --- INPUT ---
        for evento in pygame.event.get():
            if evento.type == pygame.QUIT:
                ejecutando = False

        # Obtenemos las teclas presionadas
        teclas = pygame.key.get_pressed()

        # --- UPDATE ---
        jugador.mover(teclas)
        if jugador.rect.colliderect(moneda.rect):
            puntuacion += 1  # Aumentamos la puntuación en 1
            # Movemos la moneda a una nueva posición aleatoria
            moneda.rect.x = random.randint(0, ANCHO - moneda.rect.width)
            moneda.rect.y = random.randint(0, ALTO - moneda.rect.height)




        # --- RENDER ---
        # 2. RENDERIZAR EL TEXTO
        texto_puntuacion = fuente.render(f"Puntuación: {puntuacion}", True, BLANCO)
        PANTALLA.fill(AZUL_OSCURO)
        jugador.dibujar(PANTALLA)
        moneda.dibujar(PANTALLA)
        # 3. DIBUJAR EL TEXTO EN LA PANTALLA
        PANTALLA.blit(texto_puntuacion, (10, 10))  # Lo ponemos en la esquina sup. izq.

        pygame.display.flip()



        # --- CONTROL DE TIEMPO ---
        # Le decimos al reloj que espere lo necesario para que el bucle
        # no se ejecute más de 60 veces por segundo.
        reloj.tick(60)
    pygame.quit()


### Explicación:

pygame.font.Font(None, 36): Crea un objeto de fuente. None le dice a Pygame que use su fuente básica predeterminada.

fuente.render(texto, antialias, color): Convierte nuestro string en una imagen. True activa el antialiasing para que las letras se vean más suaves.

PANTALLA.blit(imagen, (x, y)): blit es el término para "copiar píxeles" de una imagen a otra. Aquí, copiamos la imagen de nuestro texto a la pantalla principal en las coordenadas (10, 10).
